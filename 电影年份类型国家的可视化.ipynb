```python
from pyecharts.charts import Pie, Bar, TreeMap, Map, Geo,Line
from wordcloud import WordCloud, ImageColorGenerator
from pyecharts import options as opts
import matplotlib.pyplot as plt
from PIL import Image
import pandas as pd
import numpy as np
import jieba

df=pd.read_csv('Top250.csv')
df
```

```python
data_country= df['movie_type'].str.split('/', expand=True)
data_country['movie_date']=df[['movie_date']]
data_country = data_country.sort_values('movie_date', ascending=False)
dff=pd.DataFrame(columns=data_country.columns)
dff=pd.concat([data_country[data_country[0]=='剧情'], data_country[data_country[1]=='剧情'],data_country[data_country[2]=='剧情'],
           data_country[data_country[3]=='剧情'],data_country[data_country[4]=='剧情'] ],axis=0)
Bins = [1930,1940, 1950,1960, 1970,1980, 1990, 2000,2010, 2020]
Labels = ['1930-1940', '1940-1950', '1950-1960', '1960-1970', '1970-1980', '1980-1990', '1990-2000', '2000-2010', '2010-2020']
len_stage = pd.cut(dff['movie_date'], bins=Bins,labels=Labels).value_counts().sort_index()
# 获取数据
attr = len_stage.index.tolist()
v1 = len_stage.values.tolist()
# 生成折线图
bar = Line(init_opts=opts.InitOpts(width="1000px", height="400px"))
bar.add_xaxis(attr)
bar.add_yaxis("", v1)
bar.set_global_opts(title_opts=opts.TitleOpts(title="剧情类型的电影的数量的变化", pos_left="center", pos_top="18"),
                        toolbox_opts=opts.ToolboxOpts(is_show=True, feature={"saveAsImage": {}}),
                        yaxis_opts=opts.AxisOpts(splitline_opts=opts.SplitLineOpts(is_show=True)))
bar.set_series_opts(label_opts=opts.LabelOpts(is_show=True, position="top", color="black"))
bar.render("剧情类型的电影的数量的变化.html")
```

```python
# 电影年份
df = df.sort_values('movie_date', ascending=False)
# 将数据分段
Bins = [1930,1940, 1950,1960, 1970,1980, 1990, 2000,2010, 2020]
Labels = ['1930-1940', '1940-1950', '1950-1960', '1960-1970', '1970-1980', '1980-1990', '1990-2000', '2000-2010', '2010-2020']
len_stage = pd.cut(df['movie_date'], bins=Bins,labels=Labels).value_counts().sort_index()
# 获取数据
attr = len_stage.index.tolist()
v1 = len_stage.values.tolist()
# 生成柱状图
bar = Bar(init_opts=opts.InitOpts(width="1000px", height="400px"))
bar.add_xaxis(attr)
bar.add_yaxis("", v1)
bar.set_global_opts(title_opts=opts.TitleOpts(title="电影的年份分布情况", pos_left="center", pos_top="18"),
                        toolbox_opts=opts.ToolboxOpts(is_show=True, feature={"saveAsImage": {}}),
                        yaxis_opts=opts.AxisOpts(splitline_opts=opts.SplitLineOpts(is_show=True)))
bar.set_series_opts(label_opts=opts.LabelOpts(is_show=True, position="top", color="black"))
bar.render("电影的年份分布情况.html")
```

```python
# 电影类型
data_country= df['movie_type'].str.split('/', expand=True)
country_list=[]
for i in data_country.values:
    for j in i:
        if j!=None:
            country_list.append(j)
movie_country_count=pd.DataFrame(country_list)[0].value_counts()
movie_country_count=pd.DataFrame(movie_country_count).reset_index()
movie_country_count.columns=['movie_country','count']

# 饼图数据
attr = movie_country_count['movie_country']
v1 = movie_country_count['count']

# 初始化配置
pie = Pie(init_opts=opts.InitOpts(width="1800px", height="1400px"))
# 添加数据，设置半径
pie.add("", [list(z) for z in zip(attr, v1)], radius=["40%", "75%"])
# 设置全局配置项，标题、图例、工具箱(下载图片)
pie.set_global_opts(title_opts=opts.TitleOpts(title="电影的类型分布情况", pos_left="center", pos_top="top"),
                        legend_opts=opts.LegendOpts(orient="vertical", pos_left="left"),
                        toolbox_opts=opts.ToolboxOpts(is_show=True, feature={"saveAsImage": {}}))
# 设置系列配置项，标签样式
pie.set_series_opts(label_opts=opts.LabelOpts(is_show=True, formatter="{b}:{d}%"))
pie.render("电影的类型分布情况.html")
movie_country_count
```

```python
# 每个年份出现最多的电影类型
data_country= df['movie_type'].str.split('/', expand=True)
data_country['movie_date']=df['movie_date']
type_list=[]
for i in data_country.movie_date.unique():
    df1=data_country[data_country.movie_date==i]
    country_list=[]
    for i in df1[[0,1,2,3,4]].values:
        for j in i:
            if j!=None:
                country_list.append(j)
                
    df2=pd.DataFrame(country_list) 
#     print(df2[0].value_counts())
    max_type=df2[0].value_counts().index[0]
    type_list.append(max_type)

df3=pd.DataFrame({'movie_date':list(data_country.movie_date.unique()),'max_moive_type':type_list})
df3=df3.set_index('movie_date')
df3.to_csv('C:Desktop/res.csv')
```
